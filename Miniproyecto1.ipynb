En la primera figura podemos ver dos gráficos construidos con la librería matplotlib, el primero es un gráfico de torta del porcentaje de días de la semana en que se contactaron a los clientes. En éste gráfico se puede observar que los datos de tipo NAN corresponde a un 16% de los datos totales. çel segundo es un gréfico de barras con los datos de nivel de educación.

El la segunda figura se crearon dos histogramas. El primer histograma es de las edades que tienen los individuos contactados, estas edades fluctuan entre 27 y 57 años. El segundo es un histograma de la duración del contacto. Ambos gráficos se realizan usando la función histplot() de la librería seaborn.

La figura tres muestra dos gráficos, el primero un box plot para la variable 'emp.var.rate' y el segundo un gráfico de violín para la variable cons.price.idx. Ambas columnas son diferentes en sus escala, acá se ve la necesidad de normalizar opara hacer las variables "comparables".  

```python
categorias = dfA['day_of_week']
conteo_categorias = {}
for categoria in categorias:
    conteo_categorias[categoria] = conteo_categorias.get(categoria, 0) + 1
categ = list(conteo_categorias.keys())
cantidades = list(conteo_categorias.values())

educacion_categorias = dfA['education'].value_counts().index
educacion_valores = dfA['education'].value_counts().values

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].pie(cantidades, labels=categ, autopct='%1.1f%%', 
           colors=['lightblue', 'lightgreen', 'pink', 'orange', 'lightcoral', 'lightgrey'])
axs[0].set_title('Distribución de Días de la semana')

axs[1].bar(educacion_categorias, educacion_valores, color='green')
axs[1].set_title('Distribución de Educación')
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(dfB["age"], kde=False, color='purple', ax=axes[0])
axes[0].set_xlabel('Edades')
axes[0].set_ylabel('Frecuencia')
axes[0].set_title('Histograma con edades')
sns.histplot(dfB["duration"], kde=False, color='blue', ax=axes[1])
axes[1].set_xlabel('Duración')
axes[1].set_ylabel('Frecuencia')
axes[1].set_title('Histograma con duración')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(data=dfC, y='emp.var.rate', ax=axes[0])
axes[0].set_title('Box Plot de emp.var.rate')

sns.violinplot(data=dfC, y='cons.price.idx', ax=axes[1])
axes[1].set_title('Gráfico de Violín de cons.price.idx')

plt.tight_layout()

plt.show()
```

A continuación se usa la función merge de la librería pandas para unir los dataframes dfA y dfB, se observa que las columnas que se repiten son age_x y day_of_week_x, pero no son exactamente iguales ya que tienen información adicional donde en la otra hay datos NAN.

Se crean dos nuevas columnas 'age' y 'day_of_week' las cuales tienen la información completa y se eliminan las columnas utilizadas. Por último verificamos que no nos quedan datos faltantes.

Volvemos a realizar un merge para juntar el dataframe obtenido anteriormente con dfC, observamos que aca se repite la columna contact, ambas columnas tienen datos faltantes aunque no vienen expresados como NAN. Hacemos una función para juntar los datos en una nueva columna contact y eliminamos las anteriores.

```python
dfA_B = pd.merge(dfA, dfB, on="id")
dfA_B.head()
```

```python
dfA_B['age'] = dfA_B['age_x'].combine_first(dfA_B['age_y'])
dfA_B['day_of_week'] = dfA_B['day_of_week_x'].combine_first(dfA_B['day_of_week_y'])
dfA_B.head()
```

```python
pd.isna(dfA_B.loc[:,['age', 'age_x','age_y']]).sum()
```

```python
pd.isna(dfA_B.loc[:,['day_of_week', 'day_of_week_x','day_of_week_y']]).sum()
```

```python
dfA_B.drop(columns=['age_x', 'age_y', 'day_of_week_x', 'day_of_week_y'], inplace=True)
```

```python
dfA_B.head()
```

```python
dfA_B_C = pd.merge(dfA_B, dfC, on="id")
dfA_B_C.head()
```

```python
dfA_B_C.loc[:,['contact_x','contact_y']].describe()
```

```python
dfA_B_C['contact_x'].unique()
```

```python
def extraer_datos_no_faltantes(row):
    if row['contact_x'] != '-' and row['contact_y'] != 'NO INFO':
        return row['contact_x']
    elif row['contact_y'] != 'NO INFO':
        return row['contact_y']
    else:
        return row['contact_x']

dfA_B_C['contact'] = dfA_B_C.apply(extraer_datos_no_faltantes, axis=1)
dfA_B_C.drop(columns=['contact_x', 'contact_y'], inplace=True)
dfA_B_C.head()
```

```python
pd.isna(dfA_B_C).sum()
```

## 4.- Transformar a formato pedido en tabla output.csv, el resultado debe quedar igual a la tabla presentada en el archivo output.csv
## Normalizar los datos numéricos, explicar el tipo de normalización utilizada (recordar que es normalización de datos y no normalización de base de datos, i.e. Standard Scaler, MinMax Scaler, etc.) (30 ptos.):

Apoyarse en la documentación de Scikit-Learn para los distintos tipos de normalización de datos https://scikit-learn.org/stable/modules/classes.html#module-sklearn.preprocessing

Respuesta:
---
    
En primer lugar vamos a decidir que columnas tienen sentido ser normalizadas, ya que por ejemplo nr.employed(número de empleado) no es una variable que se requiera normalizar. A continuación normalizaremos las columnas duration, emp.var.rate, cons.price.idx, cons.conf.idx y euribor3m. El método elegido para normalizar será la estandarización para esto emplearemos la funciones preprocessing.StandardScaler() y fit_transform(). Esto lo realizaremos generando un nuevo dataframe con las columnas a transformar y luego pasamos cada columna al dataframe original.

```python
column_order = ['id', 'age', 'job', 'marital', 'education','default','housing','contact','month','day_of_week','duration','campaign','pdays','previous','poutcome','emp.var.rate', 'cons.price.idx','cons.conf.idx','euribor3m','nr.employed' ]
dfA_B_C_order = dfA_B_C.reindex(columns=column_order)
dfA_B_C_order.head()
```

```python
from sklearn import preprocessing

columns_to_normalize = ['duration', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m']

data_to_normalize = dfA_B_C_order[columns_to_normalize]

scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(data_to_normalize)

df_normalized = pd.DataFrame(scaled_data, columns=columns_to_normalize)

print(df_normalized.head())
```

```python
dfA_B_C_order['duration'] = df_normalized['duration']
dfA_B_C_order['emp.var.rate'] = df_normalized['emp.var.rate']
dfA_B_C_order['cons.price.idx'] = df_normalized['cons.price.idx']
dfA_B_C_order['cons.conf.idx'] = df_normalized['cons.conf.idx']
dfA_B_C_order['euribor3m'] = df_normalized['euribor3m']
```
